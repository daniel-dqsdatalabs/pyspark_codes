#### Map (função, sequência)

```
O uso do loop for pode tornar o código extremamente lento quando estamos processando grandes volumes de dados. Sempre que possível vale a pena evitar o seu uso. Para isso, podemos utilizar a função map() que aplica uma ação (função) a todos os elementos de uma lista e retorna uma nova lista com os elementos alterados pela função escolhida.

Como exemplo, podemos utilizar uma função preço de uma corrida de taxi, onde temos um valor fixo, que representa o valor cobrado ao entrar no taxi (R$ 2,50), mais um valor variável correpondente a quilometragem percorrida, multiplicado por uma taxa constante de (R$ 1,50).

No final, teremos a função: p(x) = 2.5 + 1.5 * x, onde x representa a quilometragem percorrida.

Desse modo, a função built-in map aplicou a função p iterativamente a cada um dos elementos da lista passada como parametro retornando um iterator (<map>). Contudo, como vamos precisar de uma lista como retorno, precisamos utilizar também a função list. 

```

#### Lambda
 
```
Podemos tornar ainda mais eficiente o exemplo da corrida de taxi, a partir de uma expressão lambda. Quando o conjunto de dados é pequeno, a diferença de tempo computacional entre utilizar funções lambda e função com a diretiva def é imperceptível. Por outro lado, quando estamos trabalhando com grandes volumes de dados (TB, por exemplo) a percepção de ganho temporal se torna evidente.

```
 
#### Reduce (função, sequência)
 
```
A função reduce recebe dois argumentos, onde o primeiro é uma função e o segundo é uma sequência de elementos (uma lista, por exemplo). 

Essa função diferentemente da função map, aplica a cada elemento da sequência a função passada como parametro até que reste apenas um elemento. As operações com a função reduce() são feitas em pares. Assim, reduce() aplica a função soma aos pares dos elementos da lista até obter o somatório total.

Essa função apresenta uma particularidade. Embora seja uma função built-in (que em geral não precisa ser importada), a função reduce() precisa ser importada de functools.

Como exemplo, vamos calcular o faturamento total do taxista a partir da lista de faturamento por corrida.


```

#### Filter (função, sequência)
 
```
Assim como as funções map() e reduce(), a função filter() também recebe como argumentos uma função predefinida e uma sequencia de dados.

A função filter() permite que possamos fazer filtros em elementos de uma sequencia de valores de maneira otimizada. Uma vez encontrado o valor baseado no filtro da função, iremos ter um retorno True da função filter().

Essa função não será muito útil no exemplo acima, entretanto, para não passar em branco vamos retornar uma lista com as distâncias pares.
```

#### Zip
 
```
A função zip() foi construída visando agregar valores de duas sequencias retornando uma tupla. Esse tipo de operação é muito útil quando estamos querendo fazer loops em mais de uma lista simultaneamente. Vale ressaltar que podemos utilizar zip() em duas sequencias com numeros diferentes de elementos. No entanto, o número de tuplas será equivalente ao número da menor sequencia de valores.

A função zip() é muito utilizada em processos de permutação de valores entre listas e dicionários.
```

#### Enumerate
 
```
a função enumerate() fornece os índices de cada valor em uma sequencia. Assim, enumerate() retorna uma tupla com o formato:
tupla(indice, valor)

Essa função recebe apenas um parametro que é exatamente a sequencia que queremos retornar o par indice valor. Quando uma sequencia tiver valores não numericos, é bem mais fácil fazer a manipulação a partir dos indices da sequencia. 

A partir dos indices podemos fazer operações lógicas e matemáticas com mais controle uma vez que muitas vezes não sabemos quais elementos estão na sequencia.

```

````
# função map

def preco_corrida_taxi(x):
  return 2.5 + 1.5 * x

# distancias percorridas em determinado dia
distancias_percorridas = [120, 100, 75, 15, 45] 
faturamento_por_corrida = list(map(preco_corrida_taxi, distancias_percorridas))
faturamento_por_corrida
````

````
# combinando funções map e lambda teremos um processamento mais eficiente

faturamento_por_corrida = list(map(lambda x: 2.5 + 1.5 * x, distancias_percorridas))
faturamento_por_corrida
````

````
# função reduce

from functools import reduce

faturamento_total = reduce(lambda x,y: x + y, faturamento_por_corrida)
faturamento_total
````

````
# função filter

def valor_par(x):
  if x % 2 == 0:
    return True
  else:
    return False

maior_distancia_percorrida = list(filter(valor_par, distancias_percorridas))
maior_distancia_percorrida
````

````
# função zip

list_1 = [22, 11]
list_2 = [1, 2, 3, 4, 5]

dict_1 = {'a': 1, 'b':2, 'c':3, 'd':4}
dict_2 = {'aa': 11, 'bb':22, 'cc':33, 'dd':44}

# usando a função zip com lists
#list(zip(list_1, list_2))

# usando a função zip com dictionaries
#list(zip(dict_1, dict_2)) # unindo os dois dicionários pelas chaves (default)
#list(zip(dict_1.values(), dict_2.values())) # unindo os dois dicionarios pelos valores
list(zip(dict_1, dict_2.values())) # unindo chaves do dict_1 com valores do dict_2 (permutação de valores)
````

````
# função enumerate

def lista_elementos_indice_menor_que(seq, k):
  for i, x in enumerate(seq):
    if i > k:
      break
    else:
      print(x) 
      
seq = ['casa', 'abelha', 'jaca', 'parque']
lista_elementos_indice_menor_que(seq, 1)
````

````
# histograma

import numpy as np
import matplotlib.pyplot as plt

num_norms = np.random.randn(1000, 1) #1000 linhas e 1 coluna

plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.hist(num_norms, bins=25)
plt.show()
````

````
# gráfico de linha 

x = [1, 2, 3, 4]
y = [10, 15, 22, 31]
plt.plot(x, y, linestyle='--', color='green', marker='s', linewidth=3.0)

# ajustar eixos 
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.axis([0, 10, 0, 40]) #[x1, x2, y1, y2]
plt.show()
````

````
# multiplos gráficos

x = np.linspace(0, 2, 100) # cria 100 pontos igualmente intervalados entre 0 e 2
y1 = x*2 # função linear
y2 = x**2 # função quadratica
y3 = x**3 # função cubica
y4 = np.sin(x) # função seno

fig, window = plt.subplots(2, 2) # gera a janela com 2 linhas e 2 colunas (plots lado a lado)

window[0,0].plot(x, y1, 'r')
window[0,1].plot(x, y2, 'g')
window[1,0].plot(x, y3, 'b')
window[1,1].plot(x, y4, 'y')
fig.legend(['Linear', 'Quadratica', 'Cubica', 'Seno'])
````

````
# gráfico de barras

x = [1, 2, 3, 4]
y = [10, 15, 22, 31]
plt.bar(x, y, width = 0.5, color = 'red')
plt.show()
````

````
# gráfico de pizza

dados = [520, 58, 150, 97, 158]
lbl = ['Alimentos', 'Transporte', 'Lazer', 'Aluguel', 'Educação']
plt.pie(dados, labels=lbl)
plt.show()
````
